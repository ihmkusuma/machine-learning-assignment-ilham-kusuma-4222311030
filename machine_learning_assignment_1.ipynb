**Function dengan Banyak Parameter (Arbitrary Arguments) (Hands On)**

Buatlah fungsi untuk mencari "***Luas Lingkaran dan Keliling Lingkaran***" dengan jari-jari 15 cm.

1. Luas Lingkaran = 3.14 * jari-jari**2
2. keliling Lingkaran = 2 * 3.14 * jari-jari

```python
# Function untuk menghitung luas lingkaran
def luas_lingkaran(r):
    luas = 3.14 * r**2
    return luas
# Function untuk menghitung keliling lingkaran
def keliling_lingkaran(r):
    keliling = 2 * 3.14 * r
    return keliling

print(luas_lingkaran(15))
print(keliling_lingkaran(15))
```

* luas_lingkaran(r) — fungsi menerima parameter r (jari-jari), lalu menghitung 3.14 * r²
* keliling_lingkaran(r) — fungsi menerima parameter r, lalu menghitung 2 * 3.14 * r
* Kedua fungsi dipanggil dengan nilai 15 sesuai soal

**Instance Variable (Hands On)**

Buatlah class Lingkaran yang memiliki atribut : jari-jari 10 cm, pi = 3.14159
dengan method Luas Lingkaran dan Keliling Lingkaran

```python
class Lingkaran:
    def __init__(self, jari_jari):
        self.jari_jari = jari_jari
        self.pi = 3.14159

    def luas(self):
        return self.pi * self.jari_jari ** 2

    def keliling(self):
        return 2 * self.pi * self.jari_jari


# Membuat objek dari kelas Lingkaran
lingkaran1 = Lingkaran(10)

# Menampilkan hasil
print(f"Jari-jari  : {lingkaran1.jari_jari} cm")
print(f"Luas       : {lingkaran1.luas()} cm²")
print(f"Keliling   : {lingkaran1.keliling()} cm")
```

**Enkapsulasi - Main Principles of OOP (Hands On)**

bagaimana jika saya ingin mengubah gender "Male" dari user Jonathan menjadi "unknown"???

```python
class User:
    def __init__(self, username, email, gender):
        self._username = username   # Enkapsulasi dengan underscore (_)
        self._email = email
        self._gender = gender

    def get_username(self):
        return self._username

    def set_username(self, username):
        self._username = username

    def get_email(self):
        return self._email

    def set_email(self, email):
        self._email = email

    def get_gender(self):
        return self._gender

    def set_gender(self, gender):
        self._gender = gender
```

```python
user = User("Jonathan Adis Pratama", "jonathan@example.com", "")

print(user.get_username())  # Mengakses melalui metode getter
user.set_username("Jo saja")  # Mengubah melalui metode setter
print(user.get_username())

print(user.get_gender())        # Mengakses gender melalui getter
user.set_gender("Unknown")      # Mengubah gender melalui setter
print(user.get_gender())        # Menampilkan gender setelah diubah
```

1. Enkapsulasi digunakan untuk menyembunyikan implementasi internal dan melindungi atribut agar tidak dapat diakses atau diubah secara langsung.
2. Atribut username, email, dan gender diberi awalan _ untuk menandakan bahwa seharusnya diakses secara terbatas.
3. Metode getter get_username digunakan untuk mengakses nilai atribut, dan setter set_username digunakan untuk mengubahnya dengan validasi atau logika tambahan.


**OOP Principles (Hands On)**

Buatlah program yang mengandung 2/3 Main Principles of OOP selain contoh diatas. Lalu presentasikan kedepan satu-satu

```python
from abc import ABC, abstractmethod

# ABSTRACTION - Class abstrak sebagai blueprint
class Karyawan(ABC):
    def __init__(self, nama, departemen):
        self._nama = nama
        self._departemen = departemen

    @abstractmethod
    def hitung_gaji(self):
        pass

    @abstractmethod
    def info_jabatan(self):
        pass

    def info(self):
        print(f"\n{'='*40}")
        print(f"  Nama       : {self._nama}")
        print(f"  Departemen : {self._departemen}")
        print(f"  Jabatan    : {self.info_jabatan()}")
        print(f"  Gaji       : Rp {self.hitung_gaji():,.0f}")
        print(f"{'='*40}")


# INHERITANCE - Subclass mewarisi class Karyawan
class CEO(Karyawan):
    def __init__(self, nama, departemen, tunjangan):
        super().__init__(nama, departemen)
        self.__tunjangan = tunjangan

    # POLYMORPHISM
    def hitung_gaji(self):
        gaji_pokok = 100000000
        return gaji_pokok + self.__tunjangan

    def info_jabatan(self):
        return "CEO"


class VP(Karyawan):
    def __init__(self, nama, departemen, level):
        super().__init__(nama, departemen)
        self.__level = level

    # POLYMORPHISM
    def hitung_gaji(self):
        gaji = {"junior": 15000000, "mid": 30000000, "senior": 60000000}
        return gaji.get(self.__level, 0)

    def info_jabatan(self):
        return f"VP ({self.__level.capitalize()})"


class Programmer(Karyawan):
    def __init__(self, nama, departemen, level):
        super().__init__(nama, departemen)
        self.__level = level

    # POLYMORPHISM
    def hitung_gaji(self):
        gaji = {"junior": 10000000, "mid": 20000000, "senior": 4000000}
        return gaji.get(self.__level, 0)

    def info_jabatan(self):
        return f"Programmer ({self.__level.capitalize()})"


# MAIN - Membuat objek dan menampilkan info
karyawan1 = CEO("Kusuma", "Leader of Company", 00000000)
karyawan2 = VP("Besli", "HRM", "senior")
karyawan3 = Programmer("Syafi", "Senior Dev", "mid")

daftar_karyawan = [karyawan1, karyawan2, karyawan3]

print("\n📋 DATA KARYAWAN PERUSAHAAN")
for karyawan in daftar_karyawan:
    karyawan.info()
```

1. **Abstraction**

Fungsi: Class Karyawan adalah blueprint (kerangka) yang tidak bisa dibuat objeknya secara langsung. @abstractmethod memaksa setiap subclass untuk wajib mengimplementasikan method hitung_gaji() dan info_jabatan(). Jika tidak diimplementasikan, program akan error.

2. **Inheritance**

Fungsi: Class CEO, VP, dan Programmer mewarisi atribut _nama dan _departemen serta method info() dari class Karyawan melalui super().__init__(). Sehingga tidak perlu menulis ulang kode yang sama di setiap subclass.

3. **Polymorphism**

Fungsi: Method hitung_gaji() dan info_jabatan() dipanggil dengan nama yang sama di semua subclass, tetapi hasilnya berbeda-beda tergantung jenis karyawannya. Ini terlihat jelas saat looping:
